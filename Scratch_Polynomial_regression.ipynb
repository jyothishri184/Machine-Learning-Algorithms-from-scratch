#POLYNOMIAL REGRESSION

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2.1, 3.8, 6.5, 9.2, 11.3])
```

```python
degree = 2
```

```python
x_poly = np.column_stack([x**i for i in range(degree+1)]) #design matrix #since range is explicit degree +11
x_poly
```

```python
weights = np.zeros(degree+1)
weights
```

```python
learning_rate = 0.001 #steps - give small small steps
num_iterations = 10000
```

```python
for i in range(num_iterations):
  #calculate predictions
  predictions = np.dot(x_poly,weights) #y = mx+c
  print(predictions)
  #calculate error
  error = predictions - y
  print(error)
  #update the weights (coefficients) using gradient descent
  gradient = np.dot(x_poly.T, error) /len(x)
  print(gradient)
  weights -= learning_rate * gradient
  print(f'weights for iteration {i}: {weights}')
```

```python
def predict(x,weights):
  x_poly = np.array([x**i for i in range(degree+1)])
  return np.dot(x_poly,weights)  #creating for a new input
```

```python
plt.scatter(x,y,label="DATA POINTS")
x_range = np.linspace(min(x),max(x),100)
y_pred = [predict(x,weights) for x in x_range]
plt.plot(x_range, y_pred, color='r', label="POLYNOMIAL REGRESSION")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
