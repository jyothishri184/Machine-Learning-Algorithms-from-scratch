#**Overfitting and Underfitting**

Overfitting : Train-good Test-poor performance

Underfitting : Train,Test - poor performance

#**Regularisation techniques to avoid overfitting**

**Lasso  Regression** - This method performs L1 regularization.

absolute value

(REFER FORMULA)

Diamond shape - reach zero sometimes (better)



**Ridge Regression** - This method performs L2 regularization.

squared value

(REFER FORMULA)

Circular Shape - error doent reach zero

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2.1,3.8,6.5,9.2,11.3])
```

```python
x.shape
```

```python
#design matrix
x_b = np.c_[np.ones((x.shape[0],1)),x]  # x.shape[0] = 5 then 5 times 1
x_b
#c_ Translates slice objects to concatenation along the second axis, column stack
```

```python
alpha_ridge = 1.0
alpha_lasso = 0.1
num_iterations = 1000
learning_rate = 0.01
```

```python
def lasso_regression(x,y,alpha,n_iterations,learning_rate):
   m,n = x.shape
   print(f'm {m}')
   print(f'n {n}')
   theta = np.zeros(n)
   for i in range(n_iterations):
      y_pred = x.dot(theta)
      gradient = (x.T.dot(y_pred-y)+alpha * np.sign(theta))/m  #np.sign() returns -1 when x<0 , 1 when x>0 , 0 when x==0
      theta -= learning_rate * gradient #alpha * sign(theta) - penalty
   return theta
```

```python
def ridge_regression(x,y,alpha,n_iterations,learning_rate):
   m,n = x.shape
   print(f'm {m}')
   print(f'n {n}')
   theta = np.zeros(n)
   for i in range(n_iterations):
      y_pred = x.dot(theta)
      gradient = (x.T.dot(y_pred-y)+alpha * (theta))/m  #np.sign() returns -1 when x<0 , 1 when x>0 , 0 when x==0 #alpha * (theta) - penalty
      theta -= learning_rate * gradient #alpha * (theta) - penalty
   return theta
```

```python
theta_ridge = ridge_regression(x_b,y,alpha_ridge,num_iterations,learning_rate)
theta_lasso = lasso_regression(x_b,y,alpha_lasso,num_iterations,learning_rate)
```

```python
plt.scatter(x,y,label='DATA')
```

```python
plt.scatter(x,y,label='DATA')
plt.plot(x,x_b.dot(theta_ridge),label=f'Ridge (L2) Regression',color = 'red')
plt.plot(x,x_b.dot(theta_lasso),label=f'Lasso (L1) Regression',color = 'green')
```

```python
plt.scatter(x,y,label='DATA')
plt.plot(x,x_b.dot(theta_ridge),label=f'Ridge (L2) Regression',color = 'red')
plt.plot(x,x_b.dot(theta_lasso),label=f'Lasso (L1) Regression',color = 'green')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
